```python
#libraries
import requests
import matplotlib.pyplot as plt
import time

#empty
lat = []
long = []
times = []

#tracking
for i in range(100):
    response = requests.get("http://api.open-notify.org/iss-now.json")
    data = response.json()
    
    lat.append(float(data['iss_position']['latitude']))
    long.append(float(data['iss_position']['longitude']))
    times.append(data['timestamp'])
    
    time.sleep(10)  
    
#graphing
plt.plot(long, lat, marker='o', markersize=3)
plt.title("Tracking the ISS")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
